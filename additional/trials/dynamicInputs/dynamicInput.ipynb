```python
import xarray
import rioxarray
import os
import geopandas
from shapely.geometry import mapping


geodf = geopandas.read_file("../shape_files/Lower_48_grid_fixed.shp")
```

```python
file = "../ggcmi/AgMerra/tasmin_agmerra_1980-2010.nc4"
# file = "../ggcmi/AgMerra/tasmax_agmerra_1980-2010.nc4"
# file = "../ggcmi/AgMerra/pr_agmerra_1980-2010.nc4"
# file = "/home/samay/Downloads/pr_agmerra_1980-2010.nc4"

xds = rioxarray.open_rasterio(file, decode_times=True)
```

```python
# Setting the Parameters.
print("[INFO]. Setting the parameters.")
# xds = xds.rename(x='lon', y='lat')
# xds = xds.rio.set_spatial_dims(x_dim='lon', y_dim='lat', inplace=True)
xds = xds.rio.write_crs("EPSG:4326", inplace=True)

# Clipping the file.
print("[INFO]. Clipping.")
clipped = xds.rio.clip(geodf.geometry.apply(mapping), geodf.crs, from_disk=True)
# clipped = clipped.rename(x='lon', y='lat')
# clipped = clipped.rio.set_spatial_dims(x_dim='lon', y_dim='lat', inplace=True)

clipped
```

```python
# Saving the clipped file at the location.
print("[INFO]. Saving clipped {}.".format(file))
# clipped.to_netcdf("{}".format(os.path.join("../ggcmi/AgMerra/clipped_files_v2", "tasmin.nc4")))

DF = clipped.to_dataframe()

# Closing the file.
xarray.Dataset.close(xds)

# Deleting the file from the memory.
del xds, clipped
DF
```

```python
DF = DF.reset_index()
DF = DF.dropna(how="any")
DF
```
