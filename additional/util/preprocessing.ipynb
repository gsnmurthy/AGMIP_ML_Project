```python
import os
import numpy as np
import pandas as pd
import xarray as xr
```

```python
plant_fileName = "./ggcmi/phase2_outputs/dataset/A0/plant-day/pdssat_agmerra_fullharm_plant-day_mai_global_annual_1980_2010_C360_T0_W-50_N60_A0.nc4"
maty_fileName = "./ggcmi/phase2_outputs/dataset/A0/maty-day/pdssat_agmerra_fullharm_maty-day_mai_global_annual_1980_2010_C360_T0_W-50_N60_A0.nc4"
yield_fileName = "./ggcmi/phase2_outputs/dataset/A0/yield/pdssat_agmerra_fullharm_yield_mai_global_annual_1980_2010_C360_T0_W-50_N60_A0.nc4"

plant_fileName_1 = "./ggcmi/phase2_outputs/dataset/A1/plant-day/pdssat_agmerra_fullharm_plant-day_mai_global_annual_1980_2010_C360_T0_W-50_N60_A1.nc4"
maty_fileName_1 = "./ggcmi/phase2_outputs/dataset/A1/maty-day/pdssat_agmerra_fullharm_maty-day_mai_global_annual_1980_2010_C360_T0_W-50_N60_A1.nc4"
yield_fileName_1 = "./ggcmi/phase2_outputs/dataset/A1/yield/pdssat_agmerra_fullharm_yield_mai_global_annual_1980_2010_C360_T0_W-50_N60_A1.nc4"




plantData = xr.open_dataset(plant_fileName, decode_times=False)
matyData = xr.open_dataset(maty_fileName, decode_times=False)
yieldData = xr.open_dataset(yield_fileName, decode_times=False)

plantData_1 = xr.open_dataset(plant_fileName_1, decode_times=False)
matyData_1 = xr.open_dataset(maty_fileName_1, decode_times=False)
yieldData_1 = xr.open_dataset(yield_fileName_1, decode_times=False)




units, referenceDate = plantData.time.attrs['units'].split('since')
plantData['time'] = pd.date_range(start=referenceDate, periods=plantData.sizes['time'], freq='MS')

units, referenceDate = matyData.time.attrs['units'].split('since')
matyData['time'] = pd.date_range(start=referenceDate, periods=matyData.sizes['time'], freq='MS')

units, referenceDate = yieldData.time.attrs['units'].split('since')
yieldData['time'] = pd.date_range(start=referenceDate, periods=yieldData.sizes['time'], freq='MS')




units, referenceDate = plantData_1.time.attrs['units'].split('since')
plantData_1['time'] = pd.date_range(start=referenceDate, periods=plantData_1.sizes['time'], freq='MS')

units, referenceDate = matyData_1.time.attrs['units'].split('since')
matyData_1['time'] = pd.date_range(start=referenceDate, periods=matyData_1.sizes['time'], freq='MS')

units, referenceDate = yieldData_1.time.attrs['units'].split('since')
yieldData_1['time'] = pd.date_range(start=referenceDate, periods=yieldData_1.sizes['time'], freq='MS')





# plantData['plant-day_mai'].to_dataframe().to_csv('./plantData.csv')
plantDF = plantData['plant-day_mai'].to_dataframe()
matyDF = matyData['maty-day_mai'].to_dataframe()

plantDF_1 = plantData_1['plant-day_mai'].to_dataframe()
matyDF_1 = matyData_1['maty-day_mai'].to_dataframe()





yieldDF = yieldData.yield_mai.to_dataframe()
yieldDF_1 = yieldData_1.yield_mai.to_dataframe()
# .to_csv("./temp_yield_file.csv")
# yieldDF = pd.read_csv('./temp_yield_file.csv')
# os.system("rm -rvf ./temp_yield_file.csv")

del plantData, matyData, yieldData, plantData_1, matyData_1, yieldData_1, units, referenceDate, plant_fileName, maty_fileName, plant_fileName_1, maty_fileName_1
```

```python
plant_day_mai = plantDF['plant-day_mai']
del plantDF

maty_day_mai = matyDF['maty-day_mai']
del matyDF



plant_day_mai_1 = plantDF_1['plant-day_mai']
del plantDF_1

maty_day_mai_1 = matyDF_1['maty-day_mai']
del matyDF_1
```

```python
splittedFilename = yield_fileName.split('/')[-1].split('_')

raw_CO2, raw_W, raw_T, raw_N, raw_A = splittedFilename[9], splittedFilename[11], splittedFilename[10], splittedFilename[12], splittedFilename[13]

CO2, N, A = int(raw_CO2[1:]), int(raw_N[1:]), int(raw_A[1])
T, W = raw_T.split('-'), raw_W.split('-')

if len(T) > 1:
    T = int(T[-1])
else:
    T = int(T[-1][-1])


if len(W) > 1:
    W = int(W[-1])
else:
    W = W[0].strip()
    if len(W) > 3:
        W = np.inf
    else:
        W = int(W[1:])


print("[INFO]. CO2 : ", CO2)
print("[INFO].  W  : ", W)
print("[INFO].  T : ", T)
print("[INFO].  N : ", N)
print("[INFO].  A : ", A)

del raw_CO2, raw_A, raw_N, raw_T, raw_W, splittedFilename






splittedFilename = yield_fileName_1.split('/')[-1].split('_')

raw_CO2, raw_W, raw_T, raw_N, raw_A = splittedFilename[9], splittedFilename[11], splittedFilename[10], splittedFilename[12], splittedFilename[13]

CO2_1, N_1, A_1 = int(raw_CO2[1:]), int(raw_N[1:]), int(raw_A[1])
T_1, W_1 = raw_T.split('-'), raw_W.split('-')

if len(T_1) > 1:
    T_1 = int(T_1[-1])
else:
    T_1 = int(T_1[-1][-1])


if len(W_1) > 1:
    W_1 = int(W_1[-1])
else:
    W_1 = W_1[0].strip()
    if len(W_1) > 3:
        W_1 = np.inf
    else:
        W_1 = int(W[1:])


print("[INFO]. CO2 : ", CO2_1)
print("[INFO].  W  : ", W_1)
print("[INFO].  T : ", T_1)
print("[INFO].  N : ", N_1)
print("[INFO].  A : ", A_1)

del raw_CO2, raw_A, raw_N, raw_T, raw_W, splittedFilename
```

```python
yieldDF['CO2'] = CO2
yieldDF['W'] = W
yieldDF['T'] = T
yieldDF['N'] = N
yieldDF['A'] = A

del CO2, W, T, N, A

yieldDF_1['CO2'] = CO2_1
yieldDF_1['W'] = W_1
yieldDF_1['T'] = T_1
yieldDF_1['N'] = N_1
yieldDF_1['A'] = A_1

del CO2_1, W_1, T_1, N_1, A_1
```

```python
yieldDF['plant-day'] = plant_day_mai
yieldDF['maturity-day'] = maty_day_mai



yieldDF_1['plant-day'] = plant_day_mai_1
yieldDF_1['maturity-day'] = maty_day_mai_1
```

```python
# yieldDF.yield_mai = yieldDF.yield_mai.astype('uint8')
yieldDF = yieldDF.reset_index()
yieldDF_1 = yieldDF_1.reset_index()

yieldDF.yield_mai = np.round(yieldDF.yield_mai)
yieldDF_1.yield_mai = np.round(yieldDF_1.yield_mai)

yieldDF = yieldDF.dropna(how='any')
yieldDF_1 = yieldDF_1.dropna(how='any')

yieldDF = pd.concat([yieldDF, yieldDF_1], ignore_index=True)

del yieldDF_1
yieldDF
```
